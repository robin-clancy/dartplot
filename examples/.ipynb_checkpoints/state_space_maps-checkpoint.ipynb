```python
# Edit archive and truth to point to your own data before running as this example data may not exist in the future...
# ...and if you're not using the glade filesystem it cetainly won't exist...
archive =  "/glade/derecho/scratch/chennuo/archive"
truth = "CESM2_2_BHIST_ne0ARCTICne30x4_g17_derecho_2560pes_spinup_allvars_from_20200701_2mos.004"
time = "2020-08-08-21600"
lev = 25
ds_truth = state.load_truth(archive, truth, time, lev=lev)
```

```python
exp = "FHIST_exp.1A_0001"
ds_forecast = state.load_forecast(archive, exp, time, lev=lev)
ds_output = state.load_output(archive, exp, time, lev=lev)
ds_increment = state.load_increment(ds_output, ds_forecast)
```

```python
lon = ds_truth['lon']
lat = ds_truth['lat']
```

# Map data
N.B. replace state.make_map with your own map plotting function that fits the grid type of your data

```python
state.make_map?
```

### Truth

```python
fig, ax, img = state.make_map(ds_truth['T'], lon, lat, clim_even=False, cmap="magma")
```

### Forecast error

```python
fig, ax, img = state.make_map(ds_forecast['T'] - ds_truth['T'], lon, lat, clim_even=True, cmax=10, cmap="PuOr_r")
```

### Data assimilation increment

```python
fig, ax, img = state.make_map(ds_increment['T'], lon, lat, clim_even=True, cmax=10, cmap="PuOr")
```

### Output bias

```python
fig, ax, img = state.make_map(ds_output['T'] - ds_truth['T'], lon, lat, clim_even=True, cmax=10, cmap="PuOr_r")
```

# Improvement

```python
fig, ax, img = state.make_map(np.abs(ds_forecast['T'] - ds_truth['T']) -
                              np.abs(ds_output['T'] - ds_truth['T']),
                              lon, lat, clim_even=True, cmax=5, cmap="PuOr_r")
```
