# <center>Observation space vertical profiles</center>

Example notebook for plotting observation space vertical profiles from DART filter diagnostics output.

Output is generated with obs_diag script:
https://docs.dart.ucar.edu/en/latest/assimilation_code/programs/obs_diag/threed_cartesian/obs_diag.html

See obs_space_diagnostics_contents.ipynb for example of how to find options for variable, statistics and pressure level options.

```python
from dartplot import obs
```

# Plot vertical profile of a spatially and temporally averaged statistic

```python
obs.plot_profile?
```

### Plot RMSE and total spread profiles

```python
fname = 'obs_diag_output_example.nc'
vname = 'SAT_U_WIND_COMPONENT_VPguess'
fig, ax1, h1 = obs.plot_profile(fname, vname, 'rmse', region=0, legend=True, color='k')
fig, ax2, h2 = obs.plot_profile(fname, vname, 'totalspread', region=0, fig=fig, legend=True, color='r')
ax1.set_xlabel('rmse and totalspread', fontsize=16);
```

### Add number of observations and number assimilated
Requires manually specifying the legend to get it in the correct order

```python
fname = 'obs_diag_output_example.nc'
vname = 'SAT_U_WIND_COMPONENT_VPguess'
fig, ax1, h1 = obs.plot_profile(fname, vname, 'rmse', region=0, legend=True, color='k')
fig, ax2, h2 = obs.plot_profile(fname, vname, 'totalspread', region=0, fig=fig, legend=True, color='r')
fig, ax3, h3 = obs.plot_profile(fname, vname, 'Nused', region=0, fig=fig, legend=True, color='g')
fig, ax4, h4 = obs.plot_profile(fname, vname, 'Nposs', region=0, fig=fig, legend=True, color='g')
ax1.set_xlabel('rmse and totalspread', fontsize=16);
ax3.set_xlabel('\n# observations: o=possible, *=assimilated', fontsize=16)

ax1.legend([h1, h2, h3, h4],['rmse', 'totalspread', 'Nused', 'Nposs'],
           fontsize=16, bbox_to_anchor=(1, 0.5), loc='center left');
```

### Plotting additional dart quality control numbers and changing line properties

```python
fname = 'obs_diag_output_example.nc'
vname = 'RADIOSONDE_TEMPERATURE_VPguess'
fig, ax1, h1 = obs.plot_profile(fname, vname, 'rmse', region=0, legend=True, color='k')
fig, ax2, h2 = obs.plot_profile(fname, vname, 'totalspread', region=0, fig=fig, legend=True, color='r')
fig, ax3, h3 = obs.plot_profile(fname, vname, 'Nused', region=0, fig=fig, legend=True, color='g')
fig, ax4, h4 = obs.plot_profile(fname, vname, 'Nposs', region=0, fig=fig, legend=True, color='g')
fig, ax5, h5 = obs.plot_profile(fname, vname, 'N_DARTqc_6', region=0, fig=fig, legend=True, color='m')
fig, ax6, h6 = obs.plot_profile(fname, vname, 'N_DARTqc_7', region=0, fig=fig, legend=True, color='c')
ax1.set_xlabel('rmse and totalspread', fontsize=16);
ax3.set_xlabel('\n# observations: o=possible, *=assimilated, =other', fontsize=16);

h1.set_linewidth(4)
h2.set_linewidth(4)

ax1.legend([h1, h2, h3, h4, h5, h6],['rmse', 'totalspread', 'Nused', 'Nposs', 'N_DARTqc_6', 'N_DARTqc_7'],
           fontsize=16, bbox_to_anchor=(1, 0.5), loc='center left');
```
